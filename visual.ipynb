```python
df = pd.read_csv('https://raw.githubusercontent.com/melindaleung/Ames-Iowa-Housing-Dataset/master/data/ames%20iowa%20housing.csv')
df.head()
```

```python
df.describe()
```

```python
df.SalePrice.plot(kind='hist', title='SalePrice of houses');
# Из графика видно, что из общего количества (1460 домов) большинство (чуть больше 50%) по цене от 1000000$-180000$. 
#Чем дороже, тем меньше количество.Дешевых домов больше, чем дорогих                
```

```python
df.OverallCond.value_counts().sort_values(ascending = True).plot(kind='bar', rot = 1);
# Вывод: больше всего домов с общей оценкой 5 ( 800 домов). Максимум 9, минимум 1.Скорее всего на эти дома больший спрос, 
# т.е у большинства средний уровень дохода.Следующими по количеству оценки домов на 6 и 7. Скорее их покупают лбди с доходом выше среднего.
# Меньше всего домов с оценками 1,2 (дешевых) и 8,9 (дорогих).
```

```python
df.plot(kind='scatter', 
        x='OverallCond', 
        y='SalePrice', 
        title='Correlation between SalePrice and User OverallCond')
plt.show()

# Из графика видно, что  дома с оценкой 5 имеют большой разброс цен. Это подтверждает гипотезу о высоком спросе клиентов
# со средним доходом и выше среднего. Есть несколько домов с ценой выше среднего и дорогие дома. Предположительно, высокая цена
# и средняя оценка могут зависеть от района (центр), а дом старый. Или увеличенная площадь, но район не очень. Ответить точно
# на этот вопрос можно, зная критерии оценки домов. С оценкой 9 не самое дорогое и не самое дешевое жилье. Вариантов явно меньше,
# чем у 5-ки. Возможно, эти дома строят новые застройщики. Качество лучше, чтобы завоевать рынок. Или же это дома не для отдельной
# категории людей. Или у этих домов лучше всего соотносится цена-качество-параметры критериев оценки (чуть дальше от центра,
# при этом отличная планировка, качество постройки и цена). Дом за 400000$ с оценкой 2 - явно того не стоит. С оценкой 1 стоимость
# дома ниже минимальной на графике - соответствует.
```

```python
fig, axes = plt.subplots(sharey=True, figsize=(15, 5))
sns.boxplot(x="YearBuilt", y="SalePrice", 
            data=df.loc[(df.YearBuilt>2000)]);

# В 2008 году был максимальный размах цены -  случился ипотечный кризис(на эту тему есть хороший фильм "Игра на понижение")):)
# Вероятно, после кризиса был избыток предложения (т.к нет верхнего уса) и, вряд ли строили новое жилье,но некоторые
# надеялись продать дома по высокой цене.В 2006 году цена была распределена равномерно. Одинаково спрос/предложение. Выбросы близко
# к максимуму - нет сильнго разброса.
#Медианное значение цены за 2006 и 2007 год не менялось. Образовался накопительный эффект. В 2008 году медиана сильно выросла.
# За 2010 год нет данных о построенных домах
```

```python
corrmat = df.fillna(0).corr()
f, ax = plt.subplots(figsize=(12, 9))
sns.heatmap(corrmat, vmax=.8, square=True);

# На графике прекрасно просматривается зависимость всех признаков друг от друга. Например, зависимость цены и общее качество : общий материал и качество отделки
# GrLivArea : квадратные футы жилой площади над уровнем земли и TotRmsAbvGrd :Общее количество комнат выше уровня (без ванных комнат)
# Цена зависит от всех факторов в большей иои меньшей степени, т.к. нет нулевой корелляции.
```
